# Sampling

```python
import numpy as np
import seaborn as sns
```

## Uniform

```python
n = 10000
a = 3
b = 4
sample = np.random.uniform(a,b,n)
plot = sns.histplot(sample,stat='count')
```

## Normal

```python
n = 10000
m = 2
sd = 1
sample = np.random.normal(m,sd,n)
plot = sns.histplot(sample,stat='density')
```

## Exponential

```python
n = 10000
r = 2
sample = np.random.exponential(r,n)
plot = sns.histplot(sample,stat='density')
```

## Geometric

```python
n = 10000
p = 0.3
sample = np.random.geometric(p,n)
plot = sns.histplot(sample,stat='density',discrete=True,binrange=(0,25))
```

```python
n = 10000
p = 0.7
sample = np.random.geometric(p,n)
plot = sns.histplot(sample,stat='density',discrete=True,binrange=(0,25))
```

## Poisson

```python
n = 10000
r = 5
sample = np.random.poisson(r,n)
plot = sns.histplot(sample,stat='density',discrete=True)
```

## Total payout

```python
# To obtain n samples of total payout in 1 year
# Here total payout S_N = X_1 + ... + X_N
# N is assumed to be from Poisson distribution with rate = r
# X_i are assumed to be iid exponentially distributed with mean = m

# To obtain 1 sample of total payout
def SimulateOneRun(r,m):
    # To generate one sample of no. of claims in a year
    N = np.random.poisson(r,1)
    
    # To obtain N claim samples
    X = np.random.exponential(m,N)
    
    # To obtain total payout
    S = sum(X)
  
    return S

# For obtaining n samples of the total payout 
def TotalPayout(n,r,m):
  samples = [SimulateOneRun(r,m) for i in range(n)]
  return samples
```

```python
n = 10000
r = 67
m = 210
sample = TotalPayout(n,r,m)
plot = sns.histplot(sample,stat='density')
```
