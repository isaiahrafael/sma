# Evaluate $\pi$

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Evaluating pi from the area of circle inscribed in the square [-1,1] x [-1,1]

def EvaluatePi(n):
    # n = number of samples for each co-ordinate

    x = np.random.uniform(-1,1,n)
    y = np.random.uniform(-1,1,n)
    I = (x**2 + y**2 <= 1) # I=1 if x^2 + y^2 <=1, 0 otherwise
    ax = sns.scatterplot(x=x,y=y, s=10, color="b", marker="x") # plot co-ordinates

    # set axes
    ax.set(xlim=(-1, 1))
    ax.set(ylim=(-1, 1))
    ax.set_aspect('equal')
    
    circle = plt.Circle(xy=(0, 0), radius=1, color='black', fill=False) # draw circle
    ax.add_patch(circle) # add circle to plot
    plt.plot(0,0,'ko') # draw centre of circle
    plt.show()

    avg_I = sum(I)/n
    pi_estimate = 4*avg_I
    return pi_estimate
```

```python
np.random.seed(123456)
EvaluatePi(10000)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def EvaluatePi(samples, seed):
    np.random.seed(seed)
    x = np.random.uniform(-1, 1, samples)
    y = np.random.uniform(-1, 1, samples)
    inside_circle = (x**2 + y**2) <= 1
    
    ax = sns.scatterplot(x=x, y=y, s=10, color="b", marker="x")
    ax.set(xlim=(-1, 1), ylim=(-1, 1))
    ax.set_aspect('equal')
    circle = plt.Circle((0, 0), radius=1, color='black', fill=False)
    ax.add_patch(circle)
    plt.plot(0, 0, 'ko')
    plt.show()

    return (np.sum(inside_circle) / samples) * 4

def find_closest_seed(target=np.pi, samples=10000, seed_start=0, seed_end=10000):
    closest_seed = seed_start
    closest_pi = EvaluatePi(samples, closest_seed)
    min_error = abs(closest_pi - target)

    for seed in range(seed_start + 1, seed_end + 1):
        pi_estimate = EvaluatePi(samples, seed)
        error = abs(pi_estimate - target)
        if error < min_error:
            closest_seed = seed
            closest_pi = pi_estimate
            min_error = error

    return closest_seed, closest_pi, min_error

# Example usage:
closest_seed, closest_pi, error = find_closest_seed(samples=1000, seed_start=0, seed_end=10)
print(f"Closest Seed: {closest_seed}, Pi Approximation: {closest_pi}, Error: {error}")
```
