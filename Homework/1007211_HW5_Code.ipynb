```python
import numpy as np

def GenInverseTransform(n_samples):
    samples = []
    for _ in range(n_samples):
        u = np.random.uniform(0, 1)
        
        if u < 1/3:
            sample_X = 3 * u
        elif u < 2/3: 
            sample_X = 2
        else:          
            sample_X = 3 * u

        samples.append(sample_X)

    return samples
```

```python
# generate 10000 samples
n_samples = 10000
samples = GenInverseTransform(n_samples)
```

```python
# plot histogram
sns.histplot(samples,stat='density',binwidth=0.01,cumulative=True)
# plot CDF
x = [i for i in np.arange(0,3,0.01)]
def CDF_X(x):
    if x <= 0:
        u = 0
    elif x < 1:
        u = x/3
    elif x < 2:
        u = 1/3
    elif x <= 3:
        u = x/3
    else:
        u = 1
    return u

y = [CDF_X(i) for i in x]
plt.plot(x,y,color='red')
plt.show()
```
