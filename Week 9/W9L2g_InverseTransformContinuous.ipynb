# Inverse Transform Method (Continuous)

```python
import math
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# To obtain n samples based on inverse of target CDF
# where target CDF is exponential CDF with mean mu
def GenInverseTransformExponential(mu, n_samples):
    # Initialize
    samples = []

    for i in range(n_samples):
        U = np.random.uniform(0, 1)
        sample_X = -mu * np.log(1 - U) # compute the sample of X by applying the inverse CDF to U
        samples.append(sample_X) # append to list of samples
    return samples
```

```python
# generate 10000 samples
n_samples = 10000
mu = 30
samples = GenInverseTransformExponential(mu, n_samples)
```

```python
# plot histogram
sns.histplot(samples,stat='density')
# plot PDF (not plotting CDF) of exponential distribution
x = [i for i in np.arange(0.1,250,0.1)]
y = [math.exp(-t/mu)/mu for t in x]
plt.plot(x,y,color='red')
plt.show()
```
