# Inverse Transform Method (Discrete)

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def GenerateGrades(n_samples):
    X = []
    for _ in range(n_samples):
        U = np.random.uniform(0, 1)  # Generate a scalar float
        if U <= 0.10:
            grade_sample = "A"
        elif U <= 0.50:  # 0.10 + 0.40
            grade_sample = "B"
        elif U <= 0.90:  # 0.50 + 0.40
            grade_sample = "C"
        elif U <= 0.98:  # 0.90 + 0.08
            grade_sample = "D"
        else:           # Remaining 0.02
            grade_sample = "F"
        X.append(grade_sample)
    return X
```

```python
# generate 10000 samples
X=GenerateGrades(10000)
```

```python
# check grades of class
print(X[:35])
```

```python
# plot histogram
sns.countplot(x=X,order=['A','B','C','D','F'],stat='proportion')
plt.show()
```
